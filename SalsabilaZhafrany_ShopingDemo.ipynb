```python
BNoDeliveryNo = 4/9
BYesDeliveryNo = 5/9

BNoDeliveryYes = 10/41
BYesDeliveryYes = 31/41

BNoDelivery = 14/50
BYesDelivery = 36/50
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = NoBuy = BNo
# P(B) = P(DiscountYes) * P(Free Delivery Yes) * P(DayHoliday)
# P(B|A)1 = (PHoliday|No Buy) = PHolidayNo
# P(B|A)2 = (PYesDiscount|No Buy) = PYesDiscountNo
# P(B|A)3 = (PYesDelivery|No Buy) = PYesDeliveryNo
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)3 = PYesDelivery

#ditanyakan P(A|B)?
#Rumus
#P(A|B) = P(B|A)*P(A)/P(B)
```

```python
PNoBuy1 = ((PHolidayNo*PYesDiscountNo*PYesDeliveryNo)*PNo)/(PHoliday*PYesDiscount*PYesDelivery)
print(f'P(NoBuy|Holiday, YesDiscount, YesDelivery): {PNoBuy1}')
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = Buy = BNo
# P(B) = P(DiscountYes) * P(Free Delivery Yes) * P(DayHoliday)
# P(B|A)1 = (PHoliday|Buy) = PHolidayNo
# P(B|A)2 = (PYesDiscount|Buy) = PYesDiscountNo
# P(B|A)3 = (PYesDelivery|Buy) = PYesDeliveryNo
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)3 = PYesDelivery

#ditanyakan P(A|B)?
#Rumus
#P(A|B) = P(B|A)*P(A)/P(B)
```

```python
PBuy2 = ((PHolidayYes*PYesDiscountYes*PYesDeliveryYes)*PYes)/(PHoliday*PYesDiscount*PYesDelivery)
print(f'P(Buy|Holiday, YesDiscount, YesDelivery): {PBuy2}')
```
